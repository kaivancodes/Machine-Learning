```python
import statistics as sts
sample = [2.74,1.23,2.63,2.22,1.90]
print(sample)
```

```python
print(sts.variance(sample))
```

```python
print(sts.stdev([1,3,5,7,9]))
print(sts.stdev(sample))
```

```python
print(sts.mode([1,3,5,7,9]))
print(sts.mode(sample))
```
